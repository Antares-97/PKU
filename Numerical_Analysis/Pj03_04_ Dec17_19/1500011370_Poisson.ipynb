```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
# Set maximum iteration
maxIter = 100
TOL = 1e-5

# Set Dimension and delta
lenX = 6
lenY = 5 #we set it rectangular
delta = 1

# Boundary condition
Ttop = 100
Tright = 0
def Tbottom(x):
    if x<=6 and x>=0:
        return x(6-x)
    else:
        return 0

def Tleft(y):
    if y<=5 and y>=0:
        return y(5-y)
    else:
        return 0

# heat 
q = 1.5
K = 1.04
h = 0.4
k = 1/3
n = int(lenX/h)  # n = 15
m = int(lenY/k) # m = 15
```

```python
x = np.arange(n)
y = np.arange(m)

w = np.zeros((n, m))
for i in range(1, n):
    x[i] = i * h
    y[i] = i * k
```

```python
_lambda = h* h /k* k 
mu = 2*(1+k) 
l = 1
```

```python
while(l<=maxIter):
    
    z = (-pow(h, 2)*(-q/K)+y[m-1]*(5-y[m-1])+
         _lambda * w[1][m-2]+w[2][m-1])/mu
    NORM = abs(z-w[1][m-1])
    w[1][m-1] = z
    
    for i in range(2, n-1):
        
        z = (-pow(h, 2)*(-q/K)+ 0 +w[i-1][m-1]
             +w[i+1][m-1] +_lambda*w[i][m-2])/mu
        if abs(w[i][m-1]-z)>NORM:
            NORM = abs(w[i][m-1]-z)
        w[i][m-1] = z
        
    z = (-pow(h, 2)*(-q/K)+ 0 + 0 
         +w[n-2][m-1]+w[n-1][m-2] )/mu
    
    if abs(w[n-1][m-1] - z) >NORM:
        NORM = abs(w[n-1][m-1] - z)
    w[n-1][m-1] = z
    
    for j in range(m-2, 1, -1): # from m-2, .... , 2 
        
        z = (-pow(h, 2)*(-q/K)+ y[j]*(5-y[j]) +
             _lambda*w[1][j+1] + _lambda*w[1][j-1]+w[2][j] )/mu
        if abs(w[1][j] - z) > NORM:
            NORM = abs(w[1][j] - z)
        w[1][j] = z
        
        for i in range(2, n-1):
            z = (-pow(h, 2)*(-q/K)+ w[i-1][j] 
                 +_lambda*w[i][j+1] + w[i+1][j]+ _lambda*w[1][j-1] )/mu
            if abs(w[i][j] - z) > NORM : 
                NORM = abs(w[i][j] - z)
            w[i][j] = z
        
        z = (-pow(h, 2)*(-q/K)+ 0 +w[n-2][j+1] 
             + _lambda*w[n-1][j+1]+ _lambda*w[n-1][j-1] )/mu
        if abs(w[n-1][j] - z) > NORM:
            NORM =  abs(w[n-1][j] - z)
        w[n-1][j] = z
    
    z = (-pow(h, 2)*(-q/K)+ y[1]*(5-y[1]) 
         + _lambda*x[1]*(5-x[1]) + _lambda*w[1][2] + w[2][1])/mu
    if abs(w[1][1] - z) > NORM:
        NORM = abs(w[1][1] - z)
    w[1][1] = z
    
    for i in range(2, n-1):
        z = (-pow(h, 2)*(-q/K)+ _lambda*x[i]*(5-x[i]) 
             + w[i-1][1] + _lambda*w[i][2] + w[i+1][1])/mu
        if abs(w[i][1] - z) > NORM:
            NORM = abs(w[i][1] - z)
        w[i][1] = z
    
    z = (-pow(h, 2)*(-q/K)+ 0 + _lambda*x[n-1]*(6-x[n-1]) 
         + w[n-2][1] + _lambda*w[n-1][2])/mu
    if abs(w[n-1][1] - z) > NORM: 
        NORM = abs(w[i-1][1] - z)
    w[n-1][1] = z
    
    if NORM <= TOL:
        
        for i in range(1, n):
            for j in range(1, m):
                print(x[i], y[j], w[i][j])
        
        break
    
    l = l+1
```
